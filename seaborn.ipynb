```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Example data
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015],
    'Value': [200, 170, 100, 200, 400, 370]
}

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)

# Create a line chart
sns.set_theme(style="darkgrid")
sns.lineplot(x='Year', y='Value', data=df, marker='o', linestyle='--')
plt.figure(figsize=(16, 10), dpi=300)
```

```python
# Example data with a third variable
data = {
    'X': [1, 2, 3, 4, 5, 6],
    'Y': [5, 6, 7, 8, 7, 6],
    'Category': ['A', 'B', 'A', 'B', 'A', 'B']
}

df = pd.DataFrame(data)

# Scatter plot with hue
sns.scatterplot(x='X', y='Y', hue='Category', data=df, marker='D')
```

```python
# Example data
data = {
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Value': [10, 15, 14, 8, 12, 15, 10, 18, 16]
}

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)

# Create a box plot
sns.boxplot(x='Category', y='Value',hue='Category', data=df,palette='Set2')
```

```python
# Example data with an additional categorical variable
data = {
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Value': [10, 15, 14, 8, 12, 15, 10, 18, 16],
    'Group': ['G1', 'G2', 'G1', 'G2', 'G1', 'G2', 'G1', 'G2', 'G1']
}

df = pd.DataFrame(data)

# Violin plot with split
sns.violinplot(x='Category', y='Value', hue='Group', data=df, palette='pastel',inner='quartile')
```

```python
# Example data with an additional categorical variable
data = {
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Value': [10, 15, 14, 8, 12, 15, 10, 18, 16],
    'Group': ['G1', 'G2', 'G1', 'G2', 'G1', 'G2', 'G1', 'G2', 'G1']
}

df = pd.DataFrame(data)

# Swarm plot with hue
sns.swarmplot(x='Category', y='Value', hue='Group', data=df, palette='coolwarm')
```

```python
sns.barplot(x='Category', y='Value',hue='Category', data=df, palette='pastel', errorbar=None)
```

```python

# Example data
data = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Subcategory': ['X', 'Y', 'X', 'Y', 'X', 'Y'],
    'Value': [5, 10, 15, 7, 12, 9]
}

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)

# Create a point plot
sns.pointplot(x='Category', y='Value', hue='Subcategory', data=df, palette='coolwarm',markers=['o', 's'], linestyles=['-', '--'])
```

```python
# Example data with an additional categorical variable
data = {
    'Category': ['A', 'B', 'A', 'B', 'C', 'A', 'C', 'B', 'A', 'B', 'B', 'C'],
    'Subcategory': ['X', 'Y', 'X', 'Y', 'X', 'Y', 'X', 'Y', 'X', 'Y', 'Y', 'X']
}

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)
sns.countplot(x='Category', hue='Subcategory',data=df, palette='pastel',order=['C', 'B', 'A'],)
```

```python
# Generate random data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Create a KDE plot
sns.kdeplot(x='Value', data=df)
```

```python
# Example data with two different groups
data1 = np.random.normal(loc=0, scale=1, size=1000)
data2 = np.random.normal(loc=2, scale=1.5, size=1000)

# KDE plot for two distributions
sns.kdeplot(data1, color='blue', fill=True, label='Group 1')
sns.kdeplot(data2, color='green', fill=True, label='Group 2')
```

```python
sns.kdeplot(x='Value', data=df, bw_adjust=0.3, fill=True)
```

```python
iris = sns.load_dataset("iris")

# Set up the figure
f, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect("equal")

# Draw a contour plot to represent each bivariate density
sns.kdeplot(
    data=iris.query("species != 'versicolor'"),
    x="sepal_width",
    y="sepal_length",
    hue="species",
    thresh=.1,
)
```
